### Choice of Model: K Nearest Neighbours Regression

K Nearest Neighbors Regression can be used in cases where data labels are continuous rather than discrete which is the case for predicting BoxOffice using Budget (and runtime later on). K Nearest Neighbors is also one of the simplest algorithm but effective as well.

The K Nearest Neighbors Regressor assumes that similar things exists in close proximity. It works by finding the distances between the predictor and selected K number of nearest neighbors closest to the data and averaging the values of the neighbors to generate a prediction.

```python
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.model_selection import train_test_split

# split the train and test data set
X_train, X_test, y_train, y_test = train_test_split(budget, boxoffice, test_size = 0.25, random_state = 41)
jointdf_train = pd.concat([X_train, y_train], axis = 1).reindex(X_train.index)
jointdf_test = pd.concat([X_test, y_test], axis = 1).reindex(X_test.index)
jointdf_train = jointdf_train.sort_values(by=['budget'], ascending = True)
jointdf_test = jointdf_test.sort_values(by=['budget'], ascending = True)

X_train = pd.DataFrame(jointdf_train['budget'])
y_train = pd.DataFrame(jointdf_train['BoxOffice'])
X_test = pd.DataFrame(jointdf_test['budget'])
y_test = pd.DataFrame(jointdf_test['BoxOffice'])
```

```python
# deciding on number of nearest neighbours
n_neighbors = 10
knn = neighbors.KNeighborsRegressor(n_neighbors, weights = 'uniform')
```

```python
# fit model on training dataset
knn.fit(X_train, y_train)
```

```python
# Predicting response on train and test based on model fitted to training dataset
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)
```

```python
# plot KNN regression line for train dataset
f, axes = plt.subplots(1, 1, figsize=(16, 8))
plt.scatter(X_train, y_train)
plt.scatter(X_train, y_train_pred)
plt.plot(X_train, y_train_pred, 'r-')
plt.xlabel('Budget')
plt.ylabel('BoxOffice')
plt.show()
```

```python
# plot knn regression line for test dataset
f, axes = plt.subplots(1, 1, figsize=(16, 8))
plt.scatter(X_test, y_test)
plt.scatter(X_test, y_test_pred)
plt.plot(X_test, y_test_pred, 'r-')
plt.xlabel('Budget')
plt.ylabel('BoxOffice')
plt.show()
```

```python
# R^2 value for train dataset
print('R^2:\t', knn.score(X_train, y_train))
print()

#MSE for train and test dataset
print('MSE (Train):\t', mean_squared_error(y_train, y_train_pred))
print('MSE (Test):\t', mean_squared_error(y_test, y_test_pred))
```

### Hyperparameter tuning for KNN regressor

To find out what is the best parameter to select for the K Nearest Neighbor Regressor.

Find out what is the best K number of neighbors and if uniform or distance as weight is the better parameter.

Uniform: All points in each neighborhood are weighted equally.
Distance: weight points by the inverse of their distance, meaning the closer the neighbors, the greater the influence it will have as compared to neighbors that are further away.

```python
# implementing gridsearchcv to find best value of k
from sklearn.model_selection import GridSearchCV

# create a KNeighborsRegressor object, with uniform weights
knn = neighbors.KNeighborsRegressor()

n_neighbors = [x for x in range(1,100)]
weights = ['uniform', 'distance']

params = {'n_neighbors': n_neighbors, 'weights': weights}

model = GridSearchCV(knn, params)
model.fit(X_train,y_train)
best_param = model.best_params_
knn = neighbors.KNeighborsRegressor(best_param['n_neighbors'], 
                                    weights = best_param['weights'])
print(best_param)
knn.fit(X_train, y_train)
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)
```

```python
# R^2 values and MSE after Hyperparameter tuning
print("Best paramters for KNN Regressor:\t", best_param)
print()
print('R^2:\t', knn.score(X_train, y_train))
print()

#MSE for train and test dataset
print('MSE (Train):\t', mean_squared_error(y_train, y_train_pred))
print('MSE (Test):\t', mean_squared_error(y_test, y_test_pred))
```

### Choice of Model: Decision Tree Regressor

Decision Tree Regressor is not influenced by outliers and missing values to a fair degree and is non-parametric, meaning it does not have any assumptions about the space distribution and classifier structure.

```python
from sklearn.model_selection import train_test_split

# split the train and test data set
X_train, X_test, y_train, y_test = train_test_split(budget, boxoffice, test_size = 0.25, random_state = 41)
jointdf_train = pd.concat([X_train, y_train], axis = 1).reindex(X_train.index)
jointdf_test = pd.concat([X_test, y_test], axis = 1).reindex(X_test.index)
jointdf_train = jointdf_train.sort_values(by=['budget'], ascending = True)
jointdf_test = jointdf_test.sort_values(by=['budget'], ascending = True)

X_train = pd.DataFrame(jointdf_train['budget'])
y_train = pd.DataFrame(jointdf_train['BoxOffice'])
X_test = pd.DataFrame(jointdf_test['budget'])
y_test = pd.DataFrame(jointdf_test['BoxOffice'])
```

### Hyperparameters for Decision Tree
Max_depth (depth of the tree)  
Max_leaf_nodes (number of leaf nodes)  
Max_features (maximum number of features)  
Min_samples_leaf (minimum number of samples of leaf nodes)  
Min_samples_split (the smallest sample tree of intermediate nodes)  
Min_weight_fraction_leaf (the ratio of the sample weight of the leaf node to the total weight)  
Min_impurity_split (minimum impureness) can also be adjusted

```python
# %%script false
# implementing gridsearchcv to find best value of depth
from sklearn.model_selection import GridSearchCV

# create a DecisionTree Regressor object, with uniform weights
dec_tree = DecisionTreeRegressor()

max_depth = range(1,100)
min_samples_leaf = range(1,100)

params = {'max_depth': max_depth, 'min_samples_leaf': min_samples_leaf}

model = GridSearchCV(dec_tree, param_grid=params)
model.fit(X_train, y_train)
best_param = model.best_params_
dec_tree = DecisionTreeRegressor(max_depth = best_param['max_depth'], 
                                 min_samples_leaf = best_param['min_samples_leaf'])
dec_tree.fit(X_train, y_train)
print(best_param)
y_train_pred = dec_tree.predict(X_train)
y_test_pred = dec_tree.predict(X_test)
```

```python
# R^2 values and MSE after Hyperparameter tuning
print("Best paramters for Decision Tree Regressor:\t", best_param)
print()
print('R^2:\t', dec_tree.score(X_train, y_train))
print()

#MSE for train and test dataset
print('MSE (Train):\t', mean_squared_error(y_train, y_train_pred))
print('MSE (Test):\t', mean_squared_error(y_test, y_test_pred))
```

### Hyperparameter for polynomial regression

```python
# implementing gridsearchcv to find best value of degree
from sklearn.model_selection import GridSearchCV

# create a KNeighborsRegressor object, with uniform weights
poly_reg = make_pipeline(PolynomialFeatures(), LinearRegression())

degree = np.arange(30)

params = {'polynomialfeatures__degree': degree, 
          'linearregression__fit_intercept': [True, False], 
          'linearregression__normalize': [True, False]}

model = GridSearchCV(poly_reg, param_grid=params)
model.fit(X_train, y_train)
best_param = model.best_params_
poly_reg = make_pipeline(PolynomialFeatures(best_param['polynomialfeatures__degree']), 
                                    LinearRegression(fit_intercept = best_param['linearregression__fit_intercept'],
                                                    normalize = best_param['linearregression__normalize']))

print(best_param)
poly_reg.fit(X_train, y_train)
y_train_pred = poly_reg.predict(X_train)
y_test_pred = poly_reg.predict(X_test)
```

```python
# R^2 values and MSE after Hyperparameter tuning
print("Best paramters for Polynomial Regression:\t", best_param)
print()
print('R^2:\t', poly_reg.score(X_train, y_train))
print()

#MSE for train and test dataset
print('MSE (Train):\t', mean_squared_error(y_train, y_train_pred))
print('MSE (Test):\t', mean_squared_error(y_test, y_test_pred))
```
