```python
movieData = pd.read_csv('movies.csv')
movieData.info()
```

```python
#removing rows with NA BoxOffice
ls = list(movieData[movieData['BoxOffice'].isna()].index)
movieData = movieData.drop(ls)

# shifting movie title to the front for easier reading
firstcol = movieData.pop('title')
movieData.insert(0, 'title', firstcol)
movieData.info()
```

```python
# Changing box office to numerical data
boxoffice_list = []

for i in range(5400):
    boxoffice = movieData.iloc[i].BoxOffice
    string = re.findall(r'\d+', boxoffice)
    combine_string = ''.join(string)
    boxoffice_list.append(combine_string)
len(boxoffice_list)
```

```python
dt = {'BoxOffice': boxoffice_list}
boxoffice = pd.DataFrame(dt)
boxoffice.BoxOffice = pd.to_numeric(boxoffice.BoxOffice)
movieData = movieData.reset_index(drop=True)
movieData = movieData.drop(columns=['BoxOffice'])
```

```python
movieData = pd.concat([movieData, boxoffice], axis=1).reindex(movieData.index)
movieData.info()
```

```python
#removing rows with 0 budget
movieData = movieData.loc[movieData['budget'] != 0].reset_index(drop = True)
movieData.info()
```

```python
#removing rows with NA IMDBvotes
ls = list(movieData[movieData['imdbVotes'].isna()].index)
movieData = movieData.drop(ls)
movieData.info()
```

```python
# changing imdbVotes to numerical
imdbVotes_list = []

for i in range(4187):
    imdbVotes = movieData.iloc[i].imdbVotes
    string = re.findall(r'\d+', imdbVotes)
    combine_string = ''.join(string)
    imdbVotes_list.append(combine_string)
len(imdbVotes_list)
```

```python
movieData.info()
```

```python
dt = {'imdbVotes': imdbVotes_list}
imdbVotes = pd.DataFrame(dt)
imdbVotes.imdbVotes = pd.to_numeric(imdbVotes.imdbVotes)
movieData = movieData.reset_index(drop=True)
movieData = movieData.drop(columns=['imdbVotes'])
movieData = pd.concat([movieData, imdbVotes], axis=1).reindex(movieData.index)
movieData.info()
```
